##  Polynomial  Regression
 Polynomial regression is an advancement to regular linear regression. Linear regression, accounts for the relationship between the independent variables and the dependent variable provided the the relationship is linear. 
      However, Polynomial regression sees to account for linear and non-linear relationships.
### Concepts in Polynomial regression 
  * Polynomial degree: This is the polynomial term with the highest power.
  * Polynomial term: this is the combination of a coefficient, variable and power.
  * Over_Fitting: A situation where the model learns the noise in the data, rather than the underlying pattern. 
  * Under_fitting: A situation where the model is too simple and as a result, fails to capture the underlying pattern of the data. 

```python
y = 2x*3x 
```

```python
#import the necessarylibraries
import pandas as pd
import numpy as np
data = pd.read_csv('multiple_linear_regression_dataset.csv')
```

```python
data.head(1)
```

### Polynomial uses the regular linear model's OLS (Ordinary Least Squares)

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
#assign our variables
x = data[['age','experience']]
y = data['income']
```

```python
#create a polynomial resgression of two degree
poly = PolynomialFeatures(2, include_bias = True)
x_poly = poly.fit_transform(x)
```

```python
#check the look of our independent variables after the polnomial transformation
x_poly[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#perform train test split to our data
x_train,x_test,y_train,y_test = train_test_split(x_poly,y, random_state = 42, train_size = 0.7 )
```

```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

#### fiting our models

```python
modelSK = LinearRegression()
modelSK.fit(x_train,y_train)
```

```python
modelSM = sm.OLS(y_train,x_train)
modelSM = modelSM.fit()
```

```python
r2 = modelSK.score(x_train,y_train)
r2
```

##### Make predictions

```python
y_pred = modelSK.predict(x_test)
y_pred
```

### Merics for model analysis
* R2
* MSE (Mean Squared Error: it is used to  evaluate the performance of a regression model
* rsme: squared root of MSE

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test,y_pred)
```

```python
rsme = np.sqrt(mse)
```

```python
print(f' R2 {r2},\n MSE {mse},\n RSME {rsme}')
```

### Handling over fitting using regularizations
Regularization is the penalizing of features with high coefficients, drawing coefficients that are too high to values near zero, thereby reducing models sensitivity
* Ridge (L2 regularization): is a technique used to prevent overfitting in machine learning models, particularly in linear regression and other models that involve optimization. It works by adding a penalty to the loss function, specifically a term proportional to the square of the coefficients (weights) of the model.
* Lasso (L1 regularization): is another type of regularization technique used in regression models, specifically to prevent overfitting and improve model generalization. Lasso applies L1 regularization, which differs from Ridge regularization (L2 regularization) by adding a penalty based on the absolute values of the coefficients rather than their squared values.

```python
#import the regularization libraries
from sklearn.linear_model import Ridge,Lasso
```

```python
ridgemodel = Ridge(alpha = 2)
ridgemodel = ridgemodel.fit(x_train,y_train)
ridgemodel
```

```python
ridgemodel.coef_.round(),modelSK.coef_.round()
```

#### using Lasso(L2 regularization)

```python
lassomodel = Lasso(alpha = 4)
lassomodel = lassomodel.fit(x_train,y_train)
lassomodel
```

```python
lassomodel.coef_.round(-1),modelSK.coef_.round()
```
